```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
```

```python
completeTweetsDF = pd.read_csv('../data/finalDF.csv')
completeTweetsDF.head()
```

```python
wordHappinessDF = pd.read_csv('../data/Data_Set_S1.txt', sep='\t',on_bad_lines='skip')
wordHappinessDF.head()
```

# Wrangling

```python
# Function to get rid of words not in list.
def eliminateNonListWords(tweet, wordSet):
    if type(tweet) != str:
        return np.nan
    
    return ' '.join([w for w in tweet.split(' ') if w in wordSet])

# Word set
wordSet = set(wordHappinessDF['word'])
tweetsAllWordsInList = completeTweetsDF['tweet'].apply(lambda x: eliminateNonListWords(x,wordSet))
```

```python
# Dropping NA
tweetsAllWordsInList = tweetsAllWordsInList.dropna()

# Create a CountVectorizer object and fit it to the documents
vectorizer = CountVectorizer()
vectorizer.fit(tweetsAllWordsInList)

# Get the vocabulary and generate the word count matrix
vocab = vectorizer.get_feature_names_out()
word_count_matrix = vectorizer.transform(tweetsAllWordsInList)

# Takes around a min
```

```python
# Converting an observation into [0 0 0 ... 0 0] format
denseObs = word_count_matrix[0].toarray()[0]
```
